```python
# dependencies
import pandas as pd
from splinter import Browser
import requests
from bs4 import BeautifulSoup


```

```python
# last tweet url
tweet_url = 'https://twitter.com/MarsWxReport/status/1198749371459870722'
```

```python
# response for tweet url
tweet_response = requests.get(tweet_url)
```

```python
# mars images page
marsimages_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
```

```python
# mars images response
marsimages_response = requests.get(marsimages_url)
```

```python
# mars images page beautiful soup
marsimages_soup = BeautifulSoup(marsimages_response.content, 'html.parser')
```

```python
print(marsimages_soup.prettify())

```
